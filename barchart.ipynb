```python
import matplotlib.pyplot as plt
import numpy as np

# Training features per omic per dataset
training_features = {
    'ROSMAP': [200, 200, 200],
    'BRCA': [1000, 1000, 503]
}
datasets = list(training_features.keys())
omics = ['mRNA', 'meth', 'miRNA']
colors = ['#66c2a5', '#fc8d62', '#8da0cb']

# Bar positions
x = np.arange(len(datasets))
width = 0.6

fig, ax = plt.subplots(figsize=(8, 6))

# Plotting stacked bars
bottom = [0] * len(datasets)
for i, omic in enumerate(omics):
    values = [training_features[ds][i] for ds in datasets]
    bars = ax.bar(x, values, width, bottom=bottom, label=omic, color=colors[i])

    # Add text labels to each segment
    for j, bar in enumerate(bars):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, bottom[j] + height / 2, str(height), ha='center', va='center', fontsize=9)
        bottom[j] += height  # Update bottom for next stack

# Axis settings
ax.set_title('Stacked Bar Chart: Training Features by Omics Type')
ax.set_ylabel('Number of Training Features')
ax.set_xticks(x)
ax.set_xticklabels(datasets)
ax.legend(title='Omics Type')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

methods = ['Cosine Similarity', 'Cosine Distance', 'RBF Similarity', 'RBF Distance', 'Hybrid Distance', 'Hybrid Similarity']
acc = [0.877, 0.821, 0.849, 0.858, 0.830, 0.849]
f1 = [0.876, 0.819, 0.846, 0.860, 0.820, 0.846]
auc = [0.902, 0.869, 0.892, 0.871, 0.886, 0.881]

x = np.arange(len(methods))

plt.figure(figsize=(12, 6))
plt.plot(x, acc, marker='o', label='Accuracy')
plt.plot(x, f1, marker='s', label='F1 Score')
plt.plot(x, auc, marker='^', label='AUC')

# Vertical offsets for labels to prevent overlap
offsets = {
    'acc': 0.015,
    'f1': 0.005,
    'auc': 0.025,
}

for i in range(len(methods)):
    plt.text(x[i], acc[i] + offsets['acc'], f'{acc[i]:.3f}', ha='center', fontsize=8, color='C0')
    plt.text(x[i], f1[i] + offsets['f1'], f'{f1[i]:.3f}', ha='center', fontsize=8, color='C1')
    plt.text(x[i], auc[i] + offsets['auc'], f'{auc[i]:.3f}', ha='center', fontsize=8, color='C2')

plt.xticks(x, methods, rotation=30, ha='right')  # rotate x labels for better spacing
plt.ylim(0.75, 1)
plt.ylabel('Score')
plt.title('Performance Comparison of ML Methods (ROSMAP)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
